```python
import pandas as pd
from pathlib import Path

ROLE_PROFILE_PATH = Path("skill_gap/role_skill_profiles.csv")
```

```python
role_profiles = pd.read_csv(ROLE_PROFILE_PATH)
```

```python
def detect_skill_gap(user_skill_ids, target_role_id, importance_threshold=0.2):
    """
    user_skill_ids: set of SKxxx
    target_role_id: role ID string
    importance_threshold: minimum importance to consider
    """

    role_skills = role_profiles[
        (role_profiles["role_id"] == target_role_id) &
        (role_profiles["importance"] >= importance_threshold)
    ]

    required_skills = set(role_skills["skill_id"])
    missing_skills = required_skills - set(user_skill_ids)

    gap_details = role_skills[
        role_skills["skill_id"].isin(missing_skills)
    ].sort_values("importance", ascending=False)

    return {
        "target_role": target_role_id,
        "missing_skills": list(missing_skills),
        "gap_details": gap_details.to_dict(orient="records")
    }
```

```python
user_skills = {"SK007", "SK102", "SK041"}

gap = detect_skill_gap(
    user_skill_ids=user_skills,
    target_role_id="JR_SE"
)

gap
```
