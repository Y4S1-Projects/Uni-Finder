```python
import pandas as pd
from pathlib import Path

DATA_PATH = Path("data/processed/job_skill_vectors.csv")
OUTPUT_PATH = Path("skill_gap/role_skill_profiles.csv")
```

```python
df = pd.read_csv(DATA_PATH)

skill_cols = [c for c in df.columns if c.startswith("skill_")]
```

```python
role_profiles = []

for role_id, group in df.groupby("role_id"):
    skill_counts = group[skill_cols].sum()
    total_jobs = len(group)

    for skill, count in skill_counts.items():
        role_profiles.append({
            "role_id": role_id,
            "skill_id": skill.replace("skill_", "").upper(),
            "frequency": count,
            "importance": count / total_jobs
        })
```

```python
role_profiles_df = pd.DataFrame(role_profiles)
role_profiles_df.to_csv(OUTPUT_PATH, index=False)

print("Saved role skill profiles to:", OUTPUT_PATH)
```
