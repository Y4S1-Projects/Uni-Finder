```python
import pandas as pd
import numpy as np

df = pd.read_csv("data/processed/job_skill_vectors.csv")

skill_cols = [c for c in df.columns if c.startswith("skill_")]
roles = df["role_id"].unique()
```

```python
# How many roles each skill appears in
skill_role_presence = {}

for skill in skill_cols:
    roles_with_skill = df[df[skill] > 0]["role_id"].unique()
    skill_role_presence[skill] = len(roles_with_skill)

total_roles = len(roles)
```

```python
role_profiles = []

for role_id, group in df.groupby("role_id"):
    total_jobs = len(group)

    for skill in skill_cols:
        freq = group[skill].sum()
        if freq == 0:
            continue

        tf = freq / total_jobs
        idf = np.log(total_roles / skill_role_presence[skill])

        importance = tf * idf

        role_profiles.append({
            "role_id": role_id,
            "skill_id": skill.replace("skill_", "").upper(),
            "frequency": int(freq),
            "importance": round(importance, 6)
        })
```

```python
role_profiles_df = pd.DataFrame(role_profiles)
role_profiles_df.to_csv("skill_gap/role_skill_profiles.csv", index=False)

role_profiles_df.sort_values("importance", ascending=False).head(20)
```
