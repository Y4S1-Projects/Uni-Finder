```python
import pandas as pd
import numpy as np

df = pd.read_csv("data/processed/job_skill_vectors.csv")

skill_cols = [c for c in df.columns if c.startswith("skill_")]
all_roles = df["role_id"].unique()

len(all_roles), all_roles
```

```python
skill_role_presence = {
    skill: df[df[skill] > 0]["role_id"].nunique()
    for skill in skill_cols
}

total_roles = len(all_roles)
```

```python
role_profiles = []

for role_id in all_roles:
    role_df = df[df["role_id"] == role_id]
    total_jobs = len(role_df)

    for skill in skill_cols:
        freq = role_df[skill].sum()

        # IMPORTANT: do NOT skip zero-frequency skills
        tf = freq / total_jobs if total_jobs > 0 else 0
        idf = np.log((total_roles + 1) / (skill_role_presence[skill] + 1))
        importance = tf * idf

        role_profiles.append({
            "role_id": role_id,
            "skill_id": skill.replace("skill_", "").upper(),
            "frequency": int(freq),
            "importance": round(float(importance), 6)
        })
```

```python
role_profiles_df = pd.DataFrame(role_profiles)
role_profiles_df.to_csv(
    "skill_gap/role_skill_profiles.csv",
    index=False
)

# Verify ALL roles are present
sorted(role_profiles_df["role_id"].unique())
```
